## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # main axes
axes2 = fig.add_axes([0.2,0.5,.2,.2]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('ax1.x-axis')
axes1.set_ylabel('ax1.y-axis')
axes1.set_title('Axis 1')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('ax2.x-axis')
axes2.set_ylabel('ax2.y-axis')
axes2.set_title('Axis 2');
```

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Resultant is Empty canvas 
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object    
#fig
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(x, y, 'g*')
```
